```python
from pathlib import Path
import shutil
from sklearn.model_selection import train_test_split
```

```python
PROJECT_ROOT = Path.cwd().parent

PROCESSED_PATH = PROJECT_ROOT / "data" / "processed"

TRAIN_PATH = PROJECT_ROOT / "data" / "train"
VAL_PATH = PROJECT_ROOT / "data" / "validation"
TEST_PATH = PROJECT_ROOT / "data" / "test"

classes = [
    "normal",
    "crack",
    "broken_rail",
    "misalignment",
    "surface_wear"
]

print("Paths Loaded Successfully")
```

```python
for cls in classes:

    (TRAIN_PATH / cls).mkdir(parents=True, exist_ok=True)
    (VAL_PATH / cls).mkdir(parents=True, exist_ok=True)
    (TEST_PATH / cls).mkdir(parents=True, exist_ok=True)

print("Train / Validation / Test folders created.")
```

```python
for cls in classes:

    source_folder = PROCESSED_PATH / cls

    images = list(source_folder.glob("*.jpg"))

    train_imgs, temp_imgs = train_test_split(
        images,
        test_size=0.30,
        random_state=42
    )

    val_imgs, test_imgs = train_test_split(
        temp_imgs,
        test_size=0.50,
        random_state=42
    )

    for img in train_imgs:
        shutil.copy(img, TRAIN_PATH / cls / img.name)

    for img in val_imgs:
        shutil.copy(img, VAL_PATH / cls / img.name)

    for img in test_imgs:
        shutil.copy(img, TEST_PATH / cls / img.name)

print("Dataset split completed successfully.")
```

```python
for cls in classes:

    train_count = len(list((TRAIN_PATH / cls).glob("*.jpg")))
    val_count = len(list((VAL_PATH / cls).glob("*.jpg")))
    test_count = len(list((TEST_PATH / cls).glob("*.jpg")))

    print(f"\n{cls}")
    print(f"Train: {train_count}")
    print(f"Validation: {val_count}")
    print(f"Test: {test_count}")
```
