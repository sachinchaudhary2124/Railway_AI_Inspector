```python
# ============================================================
# IMPORT LIBRARIES
# ============================================================

from pathlib import Path
import shutil

print("Libraries Imported Successfully")
```

```python
# ============================================================
# DATASET PATHS
# ============================================================

PROJECT_ROOT = Path.cwd().parent

TRAIN_DIR = PROJECT_ROOT / "data" / "train"

SYNTHETIC_DIR = PROJECT_ROOT / "data" / "synthetic"

AUGMENTED_DIR = PROJECT_ROOT / "data" / "augmented_train"

AUGMENTED_DIR.mkdir(exist_ok=True)

print("Train Directory     :", TRAIN_DIR)
print("Synthetic Directory :", SYNTHETIC_DIR)
print("Augmented Directory :", AUGMENTED_DIR)
```

```python
# ============================================================
# CREATE CLASS FOLDERS
# ============================================================

CLASSES = [
    "broken_rail",
    "crack",
    "misalignment",
    "normal",
    "surface_wear"
]

for cls in CLASSES:
    (AUGMENTED_DIR / cls).mkdir(
        parents=True,
        exist_ok=True
    )

print("Augmented Dataset Folders Created\n")

for cls in CLASSES:
    print(AUGMENTED_DIR / cls)
```

```python
# ============================================================
# MERGE ORIGINAL + SYNTHETIC DATASET
# ============================================================

image_extensions = ["*.jpg", "*.jpeg", "*.png"]

for cls in CLASSES:

    target_folder = AUGMENTED_DIR / cls

    copied_count = 0

    print(f"\nProcessing: {cls}")

    # --------------------------------------------------------
    # COPY ORIGINAL TRAIN IMAGES
    # --------------------------------------------------------

    source_train = TRAIN_DIR / cls

    for ext in image_extensions:

        for img_path in source_train.glob(ext):

            destination = target_folder / img_path.name

            shutil.copy2(
                img_path,
                destination
            )

            copied_count += 1

    # --------------------------------------------------------
    # COPY SYNTHETIC IMAGES
    # --------------------------------------------------------

    source_synthetic = SYNTHETIC_DIR / cls

    for ext in image_extensions:

        for img_path in source_synthetic.glob(ext):

            destination = target_folder / img_path.name

            shutil.copy2(
                img_path,
                destination
            )

            copied_count += 1

    print(
        f"{cls}: {copied_count} images copied"
    )

print("\nAugmented Dataset Created Successfully")
```

```python
from pathlib import Path
import shutil

PROJECT_ROOT = Path.cwd().parent

TRAIN_DIR = PROJECT_ROOT / "data" / "train"
SYNTHETIC_DIR = PROJECT_ROOT / "data" / "synthetic"
AUGMENTED_DIR = PROJECT_ROOT / "data" / "augmented_train"

CLASSES = [
    "broken_rail",
    "crack",
    "misalignment",
    "normal",
    "surface_wear"
]

# Create folders
for cls in CLASSES:
    (AUGMENTED_DIR / cls).mkdir(parents=True, exist_ok=True)

print("Folders Created\n")

# Copy original train images
for cls in CLASSES:

    source_folder = TRAIN_DIR / cls
    target_folder = AUGMENTED_DIR / cls

    for idx, image_path in enumerate(source_folder.glob("*")):

        new_name = f"orig_{idx+1}{image_path.suffix}"

        shutil.copy(
            image_path,
            target_folder / new_name
        )

    print(f"{cls}: Original Images Copied")

print("\nOriginal Dataset Added\n")

# Copy synthetic images
for cls in CLASSES:

    source_folder = SYNTHETIC_DIR / cls
    target_folder = AUGMENTED_DIR / cls

    for idx, image_path in enumerate(source_folder.glob("*")):

        new_name = f"synthetic_{idx+1}{image_path.suffix}"

        shutil.copy(
            image_path,
            target_folder / new_name
        )

    print(f"{cls}: Synthetic Images Copied")

print("\nAugmented Dataset Rebuilt Successfully")
```
