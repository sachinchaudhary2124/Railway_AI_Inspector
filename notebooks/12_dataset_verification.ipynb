```python
# ============================================================
# DATASET PATH
# ============================================================

PROJECT_ROOT = Path.cwd().parent

AUGMENTED_DIR = PROJECT_ROOT / "data" / "augmented_train"

print("Dataset Path :", AUGMENTED_DIR)
```

```python
# ============================================================
# COUNT IMAGES PER CLASS
# ============================================================

image_extensions = ["*.jpg", "*.jpeg", "*.png"]

class_counts = {}

for class_folder in AUGMENTED_DIR.iterdir():

    if class_folder.is_dir():

        count = 0

        for ext in image_extensions:
            count += len(list(class_folder.glob(ext)))

        class_counts[class_folder.name] = count

print("Augmented Dataset Counts:\n")

for cls, count in class_counts.items():
    print(f"{cls}: {count} images")
```

```python
# ============================================================
# TOTAL IMAGE COUNT
# ============================================================

total_images = sum(class_counts.values())

print("Total Images :", total_images)
```

```python
from pathlib import Path

PROJECT_ROOT = Path.cwd().parent

TRAIN_DIR = PROJECT_ROOT / "data" / "train"
SYNTHETIC_DIR = PROJECT_ROOT / "data" / "synthetic"

CLASSES = [
    "broken_rail",
    "crack",
    "misalignment",
    "normal",
    "surface_wear"
]

image_extensions = ["*.jpg", "*.jpeg", "*.png"]

print("Original Train Dataset:\n")

for cls in CLASSES:

    count = 0

    for ext in image_extensions:
        count += len(list((TRAIN_DIR / cls).glob(ext)))

    print(f"{cls}: {count}")
```

```python
print("Synthetic Dataset:\n")

for cls in CLASSES:

    count = 0

    for ext in image_extensions:
        count += len(list((SYNTHETIC_DIR / cls).glob(ext)))

    print(f"{cls}: {count}")
```
