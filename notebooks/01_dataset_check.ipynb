```python
import os
from pathlib import Path

# Project root
PROJECT_ROOT = Path.cwd().parent

DATA_PATH = PROJECT_ROOT / "data" / "raw"

print("Dataset Location:")
print(DATA_PATH)
```

```python
classes = [
    "normal",
    "crack",
    "broken_rail",
    "misalignment",
    "surface_wear"
]

for cls in classes:
    folder = DATA_PATH / cls
    images = list(folder.glob("*"))

    print(f"{cls}: {len(images)} images")
```

```python
from PIL import Image

bad_images = []

for cls in classes:
    folder = DATA_PATH / cls

    for img_path in folder.glob("*"):

        try:
            img = Image.open(img_path)
            img.verify()

        except Exception as e:
            bad_images.append((img_path.name, str(e)))

print(f"Bad Images Found: {len(bad_images)}")

if bad_images:
    for img, err in bad_images:
        print(img, "->", err)
else:
    print("All images are readable.")
```
